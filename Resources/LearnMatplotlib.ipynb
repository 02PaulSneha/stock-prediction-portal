```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of plots

### Line plots 

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

```python
plt.plot(x, y);
plt.title('Simple line plot');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True)
```

## Bar Chart

```python
months = ['Jan', 'Feb','Mar','Apr']
revenue = [100,200,300,500]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months,revenue);
plt.title('Monthly Revenue Bar Chart');
plt.xlabel('Months');
plt.ylabel('Revenue');
```

### Histograms

```python
marks50 = np.random.randint(0, 100, 50) # marks of 50 student range 0 -100
marks50
```

```python
plt.hist(marks50);
plt.xlabel('Marks');
plt.ylabel('No. of students');
```

```python
plt.hist(marks50, bins =5);
```

## Scatter Plot

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 160, 155, 185, 145]
```

```python
plt.scatter(heights,weights);
plt.title('Heights vs. Weights Scatter Plot')
plt.xlabel('Height (inches)')
plt.ylabel('Weight(inches)');
```

## Subplots & Legends

```python
days = [1,2,3,4,5]
original_price = [10,20,25,30,40]
predicted_price = [8,21,25,28,38]
```

```python
plt.plot(days,original_price,'b-o', label ='Original Price')
plt.plot(days,predicted_price,'r-o', label ='Predicted Price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Wednesday')
plt.ylabel('Price');
```

## Subplots

##  Saving the figure

```python
# First subplot - Original Price
plt.subplot(3,1,1)
plt.plot(days,original_price,'b-o', label ='Original Price' )
plt.title('Original prices over time ')
plt.xlabel('Wednesday')
plt.ylabel('Price');
plt.legend();

# Second subplot - Predicted Price
plt.subplot(3,1,2)
plt.plot(days,predicted_price,'r-o', label ='Predicted Price' )
plt.title('Predicted prices over time ')
plt.xlabel('Wednesday')
plt.ylabel('Price');
plt.legend();

# Third subplot - Differences between Original and Predicted Price
price_difference = [o-p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.bar(days,price_difference,color = 'purple', label ='Price Differences' )
plt.title('Differences between Original and Predicted Price')
plt.xlabel('Days')
plt.ylabel('Price Difference');
plt.legend();

plt.tight_layout();
```

```python
plt.plot(days,original_price,'b-o', label ='Original Price')
plt.plot(days,predicted_price,'r-o', label ='Predicted Price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Wednesday')
plt.ylabel('Price');
plt.savefig('example_prediction.png', dpi = 300) # save the image
```
